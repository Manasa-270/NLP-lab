Tokenization

```python
from nltk.tokenize import sent_tokenize,word_tokenize
text ='Manasa get down from bus in anekal stop. she wants to take some food'
statement = sent_tokenize(text)
words = word_tokenize(text)
print(statement)
print(words)
```

```python
import spacy
nlp =spacy.load('en_core_web_sm')
text1 = 'Manasa get down from bus in anekal stop. She wants to take some food'
doc = nlp(text1)
```

Display Text and Pipeline Content

```python
print(doc)
```

```python
print(doc.to_array)
```

```python
for token in doc:
  print(token.text,token.is_alpha)
```

```python
for sent in doc.sents:
  print(sent.text)
```
